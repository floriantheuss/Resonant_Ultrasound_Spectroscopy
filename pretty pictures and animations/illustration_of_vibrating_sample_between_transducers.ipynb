```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline
```

```python
h = 7
w = 5
d = 1.5
x = w+2
y = 4



size = 6
fig, ax = plt.subplots(figsize=(size, size*(h-d)*1.6/x))
ax.axis('off')

numberofframes = 7
d1 = np.linspace(0, 2*np.pi*20, numberofframes*20+1)
change = 0.25
dall = 1.5 + change - change * np.cos(d1[:-1])
#dall = 1.5 + change * np.sin(d1)**2


transduceryt = np.array([ [h, h, h-d, h-d] for d in dall] )
transduceryb = np.array([ [-h+d, -h+d, -h, -h] for d in dall] )
transducerx = np.array([-w, w, w, -w])

samplex = np.array([0, x, 0, -x])
sampley = np.array([ [h-d, -h+d+y, -h+d, h-d-y] for d in dall ])


#box1 = ax.scatter(transducerx, transduceryt[0], c='goldenrod')
#ax.fill_between(transducerx, transduceryt[0], c='goldenrod')


ax.fill(transducerx, transduceryt[0], c='goldenrod')
ax.fill(transducerx, transduceryb[0], c='goldenrod')
ax.fill(samplex, sampley[0], c='firebrick')


def animate(i):
    ax.clear()
    ax.axis('off')
    ax.fill(transducerx, transduceryt[i], c='goldenrod')
    ax.fill(transducerx, transduceryb[i], c='goldenrod')
    ax.fill(samplex, sampley[i], c='firebrick')
    plt.show

    
    #box1.set_offsets(transducerx, transduceryt[i])
    #box1.set_offsets(transducerx, transduceryb[i])
    #sample.set_offsets(samplex, sampley[i])


    
anim = FuncAnimation(fig, animate, frames=numberofframes*20, interval=10, repeat=True)

plt.show()

#anim.save('schematic_setup.gif', writer='imagemagick', dpi=100)
```
